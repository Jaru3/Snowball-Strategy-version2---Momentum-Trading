```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

df = pd.read_csv("dataset/KRW-BTC.csv")
df_dict = {}
for i in range(len(df)):
    if df["datetime"][i][:10] in df_dict:
        df_dict[df["datetime"][i][:10]]+=1
    else:
        df_dict[df["datetime"][i][:10]]=1

sorted_dict = sorted(df_dict.items(), key= lambda item:item[1], reverse=False)


dff = df['open']
dff_array = np.array(dff)
dff_array = dff_array.reshape(-1,24)
col_name =["price0","price1","price2","price3","price4","price5","price6","price7","price8","price9","price10","price11","price12","price13","price14","price15","price16","price17","price18","price19","price20","price21","price22","price23"]
dff_df = pd.DataFrame(dff_array,columns=col_name)
df_24h = dff_df

df = df_24h
fee_rate = 0.0005
results = {}

#     (  )
start, end = 1500, 2000  #   
df = df.iloc[start:end]
initial_cash = df['price9'][start]


for col in df.columns[0:24]:  #   
    cash = initial_cash
    btc = 0
    portfolio = []

    prev_price = None
    for price in df[col].values:
        if prev_price is None:
            portfolio.append(initial_cash)
            prev_price = price
            continue

        daily_return = (price - prev_price) / prev_price

        if daily_return > 0 and cash > 0:
            btc = cash / (price * (1 + fee_rate))
            cash = 0
        elif daily_return < 0 and btc > 0:
            cash = btc * (price * (1 - fee_rate))
            btc = 0

        total = cash + btc * price
        portfolio.append(total)
        prev_price = price

    results[col] = portfolio

#   
result_df = pd.DataFrame(results)

#   price0, price9 original_price0, original_price9 
result_df['original_price0'] = df['price0'].values
result_df['original_price9'] = df['price9'].values

pd.set_option('display.float_format', '{:.0f}'.format)

df = result_df

#  (start) (end)  
start, end = 0, len(df)  #    (  )
df = df.iloc[start:end]

strategies = df.columns[:]

def calculate_cagr(data):
    start_value = data.iloc[0]
    end_value = data.iloc[-1]
    n = len(data) / 365
    return (end_value / start_value) ** (1 / n) - 1

def calculate_mdd(data):
    cumulative = data / data.iloc[0]
    peak = cumulative.cummax()
    drawdown = (cumulative - peak) / peak
    return drawdown.min()

def calculate_sharpe(data, risk_free_rate=0.01):
    daily_returns = data.pct_change().dropna()
    excess_returns = daily_returns - risk_free_rate / 365
    return excess_returns.mean() / excess_returns.std()

def calculate_var(data, alpha=0.05):
    daily_returns = data.pct_change().dropna()
    return np.percentile(daily_returns, alpha * 100)

def calculate_cvar(data, alpha=0.05):
    daily_returns = data.pct_change().dropna()
    var = np.percentile(daily_returns, alpha * 100)
    return daily_returns[daily_returns <= var].mean()

def calculate_ewma_metrics(data, lambda_):
    reversed_series = data[::-1]  #   
    
    if lambda_ > 0:
        span_value = 1 / (1 - lambda_)
        ewma_series = reversed_series.ewm(span=span_value).mean()
    else:
        ewma_series = reversed_series.rolling(window=10, min_periods=1).mean()  # =0    (SMA) 

    ewma_series = ewma_series[::-1]  #    
    cagr = calculate_cagr(ewma_series) * 10000
    mdd = calculate_mdd(ewma_series) * 10000
    sharpe = calculate_sharpe(ewma_series) * 10000
    var = calculate_var(ewma_series) * 10000
    cvar = calculate_cvar(ewma_series) * 10000
    score = cagr + sharpe - abs(mdd) - abs(var) - abs(cvar)
    return [cagr, mdd, sharpe, var, cvar, score]


evaluation = []
for strategy in strategies:
    series = df[strategy]
    cagr = calculate_cagr(series) * 10000
    mdd = calculate_mdd(series) * 10000
    sharpe = calculate_sharpe(series) * 10000
    var = calculate_var(series) * 10000
    cvar = calculate_cvar(series) * 10000
    score = cagr + sharpe - abs(mdd) - abs(var) - abs(cvar)
    evaluation.append([strategy, cagr, mdd, sharpe, var, cvar, score])

evaluation_df = pd.DataFrame(evaluation, columns=['Strategy', 'CAGR', 'MDD', 'Sharpe', 'VaR', 'CVaR', 'Score'])

ewma_results = {}
for lambda_ in [0.99999999, 0.999, 0.9981, 0.997, 0.993]:
    ewma_eval = []
    for strategy in strategies:
        series = df[strategy]
        ewma_eval.append([strategy] + calculate_ewma_metrics(series, lambda_))
    ewma_results[lambda_] = pd.DataFrame(ewma_eval, columns=['Strategy', 'CAGR', 'MDD', 'Sharpe', 'VaR', 'CVaR', 'Score'])

evaluation_df = evaluation_df.sort_values(by='Score', ascending=False)
for lambda_, df_ in ewma_results.items():
    ewma_results[lambda_] = df_.sort_values(by='Score', ascending=False)

print("\n  :")
print(evaluation_df)
for lambda_, df_ in ewma_results.items():
    print(f"\nEWMA(={lambda_})  :")
    print(df_)
```
