미국장과 그냥장의 상관관계

1. 미국오전장(선)과 미국오후장(후)의 상관관계 분석

2. 미국오전장(선)과 나머지장(후)의 상관관계 분석

3. 미국오후장(선)과 나머지장(후)의 상관관계 분석

4. 미국전체장(선)과 나머지장(후)의 상관관계 분석

5. 나머지장(선)과 미국오전장(후)의 상관관계 분석

6. 나머지장(선)과 미국오후장(후)의 상관관계 분석

7. 나머지장(선)과 미국전체장(후)의 상관관계 분석

원칙 - 미국장이 시작하고 가격흐름이 결정되며 물론 오후에 꺾일수 있지만 그 대세는 거의 유지한다. 미국장의 기조가 다음날 미국장이 시작하기 전까지 대체적으로 유지된다. OR 미국장이 시작할쯔음 바뀔 수 있다. OR 경험론적 관점에 따르면 유럽장이 시작하는 4시에 그 흐름이 꺾이거나 강해지거하기도 한다. 

1번의 가설
    미국오전장과 오후장은 약하지만 유의미한 양의 상관관계 존재

2번의 가설
    미국오전장과 나머지장은 약하지만 유의미한 양의 상관관계 존재

3번의 가설
    미국오후장과 나머지장은 강한 양의 상관관계 존재

4번의 가설
    미국전체장과 나머지장은 강한 양의 상관관계 존재

5번의 가설
    나머지장과 미국오전장은 무의미한 0에 수렴하는 상관관계존재

6번의 가설
    나머지장과 미국오후장은 무의미한 0에 수렴하는 상관관계존재

7번의 가설
    나머지장과 미국전체장은 무의미한 0에 수렴하는 상관관계존재

(나의 예측과 결과다 다르다면 유럽장의 영향력도 추가해서 따져보자, BUT 거의 영향력 없을것으로 예측된다.)
